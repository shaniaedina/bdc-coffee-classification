```python
import os
import matplotlib.pyplot as plt
import pandas as pd
from google.colab.patches import cv2_imshow
import cv2
from skimage.feature import hog
```

```python
# Load gambar
image = cv2.imread('/content/gdrive/MyDrive/datasets/coffee/test/1.jpg', cv2.IMREAD_GRAYSCALE)

# Inisialisasi detektor SIFT
sift = cv2.SIFT_create()

# Temukan dan ekstrak fitur SIFT
keypoints, descriptors = sift.detectAndCompute(image, None)

# Gambar keypoints pada gambar asli
image_with_keypoints = cv2.drawKeypoints(image, keypoints, None)

# Tampilkan gambar dengan keypoints
cv2_imshow(image_with_keypoints)
```

```python
# Load gambar
image = cv2.imread('/content/gdrive/MyDrive/datasets/coffee/test/1.jpg', cv2.IMREAD_GRAYSCALE)

# Hitung fitur HOG
features, hog_image = hog(image, pixels_per_cell=(16, 16), cells_per_block=(1, 1), block_norm='L2-Hys', visualize=True)

# Tampilkan gambar asli
plt.figure(figsize=(8, 4))
plt.subplot(121)
plt.imshow(image, cmap=plt.cm.gray)
plt.title('Gambar Asli')

# Tampilkan gambar hasil ekstraksi fitur HOG
plt.subplot(122)
plt.imshow(hog_image, cmap=plt.cm.gray)
plt.title('Ekstraksi Fitur HOG')
plt.show()
```
